```python
import pandas as pd
df=pd.read_csv("deliveries.csv")
print(df.head(5))
```

```python
df_deliveries = df.groupby('bowling_team')['total_runs'].size()
print(df_deliveries)
plt.figure(figsize=(8,8))
plt.bar(df_deliveries.index, df_deliveries.values, color='skyblue')
plt.title("Total Deliveries by Each Bowling Team")
plt.xlabel('Bowling Team')
plt.ylabel('Total Deliveries')
plt.xticks(rotation=45)
plt.show()


```

```python
df_most_runs=df.groupby('batter')['total_runs'].sum().sort_values(ascending=False).head(10)
print(df_most_runs)
plt.figure(figsize=(8,8))
plt.bar(df_most_runs.index, df_most_runs.values, color='skyblue')
plt.title("question2")
plt.xlabel("batter")
plt.ylabel("total_rus")
plt.xticks(rotation=45)
plt.show()
```

```python
# Corrected code for most wickets by bowler
df_bowlers = df.groupby('bowler')['is_wicket'].sum().sort_values(ascending=False).head(10)
print(df_bowlers)

# Plotting
plt.figure(figsize=(10,6))
plt.bar(df_bowlers.index, df_bowlers.values, color="lightblue")
plt.title("Top 10 Bowlers with the Most Wickets")
plt.xlabel("Bowler")
plt.ylabel("Most Wickets")
plt.xticks(rotation=45)
plt.show()
```

```python
# Clean up team names to remove any leading/trailing whitespaces
df['bowling_team'] = df['bowling_team'].str.strip()

# Group by unique bowling teams and sum extra runs
data_bowling = df.groupby("bowling_team")["extra_runs"].sum()

# Plotting the result
plt.figure(figsize=(8,8))
plt.bar(data_bowling.index, data_bowling.values, color="lightblue")
plt.title("Total Extra Runs Conceded by Each Team")
plt.xlabel("Bowling Team")
plt.ylabel("Extra Runs")
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Count the number of times each batsman was dismissed as run-out
run_out_counts = df[df['dismissal_kind'] == 'run-out'].groupby('batter').size().nlargest(5)

# Tabulate the results
run_out_table = pd.DataFrame({'Batsman': run_out_counts.index, 'Run-Out Count': run_out_counts.values})
print(run_out_table)

# Plot the results
plt.figure(figsize=(8, 5))
plt.bar(run_out_counts.index, run_out_counts.values, color='lightcoral')
plt.title("Top 5 Batsmen Involved in Most Run-Outs")
plt.xlabel("Batsman Name")
plt.ylabel("Run-Out Count")
plt.xticks(rotation=45)
plt.show()
```

```python
df_most_extra_runs=df.groupby('bowler')['extra_runs'].sum().head(5)
plt.title("question 6")
plt.bar(df_most_extra_runs.index,df_most_extra_runs.values,color="lightblue")
plt.xlabel('bowler')
plt.ylabel('most extra runs')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Count the number of dismissals per type
df_most_dismissal_kind = df['dismissal_kind'].value_counts()

# Select the top 5 most common dismissal types
top_dismissal_kind = df_most_dismissal_kind.head(5)

# Plot
plt.figure(figsize=(8, 5))
plt.bar(top_dismissal_kind.index, top_dismissal_kind.values, color='green')
plt.title("Top 5 Dismissal Types")
plt.xlabel('Dismissal Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Group by 'over' and sum 'total_runs'
df_most_total_runs = df.groupby('over')['total_runs'].sum().nlargest(5)

# Plot
plt.figure(figsize=(8, 5))
plt.bar(df_most_total_runs.index, df_most_total_runs.values, color='blue')
plt.title("Top 5 Overs with Most Total Runs")
plt.xlabel("Over")
plt.ylabel("Total Runs")
plt.xticks(rotation=0)  # Keeping it readable
plt.show()
```

```python
import matplotlib.pyplot as plt

# Filter data for powerplay (overs 1-6)
powerplay_df = df[df['over'].between(1, 6)]

# Aggregate total runs scored in the powerplay
powerplay_runs = powerplay_df.groupby('over')['total_runs'].sum()

# Plot powerplay performance
plt.figure(figsize=(8, 5))
plt.bar(powerplay_runs.index, powerplay_runs.values, color='orange')
plt.title("Powerplay Performance (First 6 Overs)")
plt.xlabel("Over")
plt.ylabel("Total Runs")
plt.xticks(range(1, 7))  # Ensures x-axis has proper ticks
plt.show()
```
